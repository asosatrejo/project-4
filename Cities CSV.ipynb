```python
# Imports
import pandas as pd
```

```python
# Read CSV file
aqi_df = pd.read_csv('annual_aqi_by_cbsa_2022.csv')

# Display first 5 rows
aqi_df.head()
```

```python
# Split CBSA column into 'City' and 'State'
aqi_df[['City', 'State']] = aqi_df['CBSA'].str.split(',', expand=True)

# Drop CBSA column
aqi_df.drop('CBSA', axis=1, inplace=True)

# Display first 5 rows
aqi_df
```

```python
# Obtain 'North Carolina' data
nc_df = aqi_df.loc[aqi_df['State'] == ' NC']

# Display first 5 rows
nc_df.head()
```
