```python
import pandas as pd
```

```python
# Create Data Frame
annual_aqi_df = pd.DataFrame()

# Obatin csv files
csv_files = []
for x in range(0, 10):
    name = f"Resources/annual_aqi_by_county_200{x}.csv"
    csv_files.append(name)

for x in range(10, 23):
    name = f"Resources/annual_aqi_by_county_20{x}.csv"
    csv_files.append(name)
    
for x in range(80, 100):
    name = f"Resources/annual_aqi_by_county_19{x}.csv"
    csv_files.append(name)

# Append all files together
for file in csv_files:
            df_temp = pd.read_csv(file)
            annual_aqi_df = annual_aqi_df.append(df_temp, ignore_index=True)
```

```python
# View Data Frame
annual_aqi_df
```

```python
# Set 'State' to index
annual_aqi_df = annual_aqi_df.set_index("State")

annual_aqi_df.head()
```

```python
# Obtain NC data
nc_aqi_df = annual_aqi_df.loc[["North Carolina"]]
nc_aqi_df
```

```python
# Set 'County' to index
nc_aqi_df = nc_aqi_df.set_index("County")
nc_aqi_df
```

```python
# Obtain data for 'Mecklenburg', 'Lincoln', 'Gaston', 'Cabarrus', 'Union'
nc_aqi_df = nc_aqi_df.loc[["Buncombe", "Mecklenburg", "Wake", "New Hanover", "Cumberland", "Pitt"]]
nc_aqi_df
```

```python
# Find duplicates
nc_aqi_df[nc_aqi_df.index.duplicated()]

# Drop duplicates
nc_aqi_df = nc_aqi_df.drop_duplicates()
nc_aqi_df
```

```python
# Print Data Types
nc_aqi_df.dtypes
```

```python
# Count unique values in 'County'
pd.Index.value_counts(nc_aqi_df.index)
```

```python
# Obtain 'New Hanover' data
new_hanover_aqi_df = nc_aqi_df.loc[["New Hanover"]]

# Sort by 'Year' in ascending order
new_hanover_aqi_df = new_hanover_aqi_df.sort_values(by=["Year"], ascending=True)

# Find missing values
y = 1980

for x in range(0, len(new_hanover_aqi_df["Year"])):
    if(new_hanover_aqi_df["Year"][x] != y):
        print(f"Missing: {y}")
        y = y + 1
    y = y+1
```

```python
# Print Rows with null values in 'State'
pd.Index.value_counts(nc_aqi_df.index)
```

```python
# Create rows for missing years for 'New Hanover'
new_hanover_aqi_df = new_hanover_aqi_df.append({"Year": 1983}, ignore_index=True)
new_hanover_aqi_df = new_hanover_aqi_df.append({"Year": 1986}, ignore_index=True)

# Sort by 'Year' in ascending order
new_hanover_aqi_df = new_hanover_aqi_df.sort_values(by=["Year"], ascending=True)

# Use data imputation to fill in missing values with the previous year's value
new_hanover_aqi_df = new_hanover_aqi_df.fillna(method="ffill")

# Add 'County' column with value 'New Hanover'
new_hanover_aqi_df["County"] = "New Hanover"

# Chnage index to 'New Hanover'
new_hanover_aqi_df = new_hanover_aqi_df.set_index("County")
new_hanover_aqi_df
```

```python
# Append the 'New Hanover' data to the 'North Carolina' data
nc_aqi_df = nc_aqi_df.append(new_hanover_aqi_df, ignore_index=False)

# Find duplicates
nc_aqi_df[nc_aqi_df.index.duplicated()]

# Drop duplicates
nc_aqi_df = nc_aqi_df.drop_duplicates()
nc_aqi_df

nc_aqi_df
```

```python
# Obtain 'Cumberland' data
cumberland_aqi_df = nc_aqi_df.loc[["Cumberland"]]

# Sort by 'Year' in ascending order
cumberland_aqi_df = cumberland_aqi_df.sort_values(by=["Year"], ascending=True)

# Find missing values
y = 1980

for x in range(0, len(cumberland_aqi_df["Year"])):
    if(cumberland_aqi_df["Year"][x] != y):
        print(f"Missing: {y}")
        y = y + 1
    y = y+1
```

```python
# Create rows for missing years
cumberland_aqi_df = cumberland_aqi_df.append({"Year": 1980}, ignore_index=True)
cumberland_aqi_df = cumberland_aqi_df.append({"Year": 1984}, ignore_index=True)

# Sort by 'Year' in ascending order
cumberland_aqi_df = cumberland_aqi_df.sort_values(by=["Year"], ascending=True)

# Use data imputation to fill in missing values with the previous year's value
cumberland_aqi_df = cumberland_aqi_df.fillna(method="ffill")

# Use data imputation to fill in missing values with the next year's value
cumberland_aqi_df = cumberland_aqi_df.fillna(method="bfill")

# Add 'County' column with value 'Cumberland'
cumberland_aqi_df["County"] = "Cumberland"

# Chnage index to 'New Hanover'
cumberland_aqi_df = cumberland_aqi_df.set_index("County")
cumberland_aqi_df


cumberland_aqi_df
```

```python
# Append the 'Cumberland' data to the 'North Carolina' data
nc_aqi_df = nc_aqi_df.append(cumberland_aqi_df, ignore_index=False)

# Find duplicates
nc_aqi_df[nc_aqi_df.index.duplicated()]

# Drop duplicates
nc_aqi_df = nc_aqi_df.drop_duplicates()
nc_aqi_df
```
