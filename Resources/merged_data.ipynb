```python
import pandas as pd
import numpy as np

```

```python
# Load the provided CSV files
file_path_aqi = '/Users/nataliiabondarenko/Desktop/GitHub/project-4/Resources/nc2023_aqi_df.csv'
file_path_cleaned_data = '/Users/nataliiabondarenko/Desktop/GitHub/project-4/Resources/cleaned_data.csv'
```

```python

# Reading the data from the files
df_aqi = pd.read_csv(file_path_aqi)
df_cleaned_data = pd.read_csv(file_path_cleaned_data)

# Displaying the first few rows of each dataframe to understand their structure
df_aqi_head = df_aqi.head()
df_cleaned_data_head = df_cleaned_data.head()

df_aqi_head, df_cleaned_data_head
```

```python
# Checking the unique values in the 'County' column of both datasets
unique_counties_aqi = df_aqi['Unnamed: 0'].unique()  # Assuming the 'Unnamed: 0' column is the county name in df_aqi
unique_counties_cleaned_data = df_cleaned_data['County'].unique()

# Displaying the unique values
unique_counties_aqi, unique_counties_cleaned_data
```

```python
# Renaming the 'Unnamed: 0' column in df_aqi to 'County' for consistency
df_aqi.rename(columns={'Unnamed: 0': 'County'}, inplace=True)

# Merging the two datasets on the 'County' column
merged_df = pd.merge(df_cleaned_data, df_aqi, on='County', how='inner')

# Displaying the first few rows of the merged dataset
merged_df_head = merged_df.head()
merged_df_head
```

```python
# Saving the merged dataset to a CSV file
merged_df.to_csv('/Users/nataliiabondarenko/Desktop/GitHub/project-4/Resources/merged_data.csv', index=False)

# Saving the merged dataset to a JSON file
merged_df.to_json('/Users/nataliiabondarenko/Desktop/GitHub/project-4/Resources/merged_data.json', orient='records', date_format='epoch', double_precision=10, force_ascii=True, date_unit='ms', default_handler=None)
```
