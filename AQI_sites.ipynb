```python
# AQI Sites
# Imports
import pandas as pd
```

```python
# Read CSV file
sites_df = pd.read_csv('aqs_sites.csv')
sites_df.head()
```

```python
# Find 'North Carolina' in the 'state' column
nc_sites_df = sites_df.loc[sites_df['State Name'] == 'North Carolina']
nc_sites_df.head()
```

```python
# Find 'Mecklenburg', 'Cabarrus', 'Lincoln', 'Gaston', and 'Union' in the 'county' column
counties = ['Mecklenburg', 'Cabarrus', 'Lincoln', 'Gaston', 'Union']
nc_counties_df = nc_sites_df.loc[nc_sites_df['County Name'].isin(counties)]
nc_counties_df.head()
```

```python
# Drop the 'State Code', 'County Code', 'Datum', 'Met Site State Code', 'Met Site County Code', 'Met Site Site Number', 'Met Site Type', 
# 'Met Site Distance', 'Met Site Direction', 'GMT Offset', 'Owning Agency', 'State Name', 'CBSA Name', and 'Tribe Name' column
nc_counties_df = nc_counties_df.drop(columns=['State Code', 'County Code', 'Datum', 'Met Site State Code', 'Met Site County Code',
                                                'Met Site Site Number', 'Met Site Type', 'Met Site Distance', 'Met Site Direction',
                                                'GMT Offset', 'Owning Agency', 'State Name', 'CBSA Name', 'Tribe Name'])
nc_counties_df.head()
```

```python
# Check for duplicate values
nc_counties_df.duplicated().sum()

# Drop duplicate values
nc_counties_df = nc_counties_df.drop_duplicates()
nc_counties_df.head()
```

```python
# Check for null values
nc_counties_df.isnull().sum()
```

```python
# Check data types
nc_counties_df.dtypes
```

```python
# Change Data Types
# Change Zip Code data type to string
nc_counties_df['Zip Code'] = nc_counties_df['Zip Code'].astype(str)

# Change Site Established Date data type to datetime
nc_counties_df['Site Established Date'] = pd.to_datetime(nc_counties_df['Site Established Date'])

# Change Site Closed Date data type to datetime
nc_counties_df['Site Closed Date'] = pd.to_datetime(nc_counties_df['Site Closed Date'])

# Change Exraction Date data type to datetime
nc_counties_df['Extraction Date'] = pd.to_datetime(nc_counties_df['Extraction Date'])
```

```python
# Change Null Values
# Change 'Land Use' null values to 'Unknown'
nc_counties_df['Land Use'] = nc_counties_df['Land Use'].fillna('Unknown')

# Change 'Location Type' null values to 'Unknown'
nc_counties_df['Location Setting'] = nc_counties_df['Location Setting'].fillna('Unknown')

# Change 'Site Closed Date' null values to 'Open'
nc_counties_df['Site Closed Date'] = nc_counties_df['Site Closed Date'].fillna('Open')

# Change 'Local Site Name' null values to 'Unknown'
nc_counties_df['Local Site Name'] = nc_counties_df['Local Site Name'].fillna('Unknown')

# Change 'Zip Code' null values to '00000'
nc_counties_df['Zip Code'] = nc_counties_df['Zip Code'].fillna('00000')


# Display DataFrame
nc_counties_df
```

```python
# Check for null values
nc_counties_df.isnull().sum()
```

```python
# Rename the 'County Name' column to 'County'
nc_counties_df = nc_counties_df.rename(columns={'County Name': 'County'})
nc_counties_df.head()
```

```python
# Export to CSV file
nc_counties_df.to_csv('nc_aqi_sites.csv', index=False)
```
